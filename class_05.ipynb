# 1. Matplotlob

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("整理好的租屋資料.xlsx")
```

```python
df.head(2)
```

```python
df.info()
```

長條圖 Bar chart

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
plt.bar(x,y)
```

```python
plt.show()
```

顯示中文

mac版本

```python
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"/System/Library/Fonts/STHeiti Light.ttc")
```

windows版本

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
plt.bar(x,y)
plt.xticks(fontproperties=font,size=10)
```

```python
plt.show()
```

# 2. 設定圖片格式

圖片顏色

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
plt.bar(x,y,color="red")
plt.xticks(fontproperties=font,size=10)
plt.show()
```

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
color = ['r','g','b','c','m','y','k','w']
plt.bar(x,y,color=color)
plt.xticks(fontproperties=font,size=10)
plt.show()
```

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
color = ['#FF3333','#FFAA33','#FFFF00','c','m','y','k','w']
plt.bar(x,y,color=color)
plt.xticks(fontproperties=font,size=10)
plt.show()
```

圖片標題、標籤

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
plt.bar(x,y)
plt.xticks(fontproperties=font,size=10)
plt.title("各縣市統計數量",fontproperties=font,size=12)
plt.xlabel("city")
plt.ylabel("number")
plt.show()
```

x,y軸範圍

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
plt.bar(x,y)
plt.xticks(fontproperties=font,size=10)
plt.title("各縣市統計數量",fontproperties=font,size=12)
plt.xlabel("city")
plt.ylabel("number")
plt.ylim(0,30)
plt.show()
```

圖片大小

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
plt.bar(x,y)
plt.xticks(fontproperties=font,size=10)
plt.title("各縣市統計數量",fontproperties=font,size=12)
plt.xlabel("city")
plt.ylabel("number")
plt.gcf().set_size_inches(10,5)
plt.show()
```

圖片格線

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
plt.bar(x,y,label="real")
plt.xticks(fontproperties=font,size=10)
plt.title("各縣市統計數量",fontproperties=font,size=12)
plt.xlabel("city")
plt.ylabel("number")
plt.grid(True)
plt.show()
```

儲存圖片

```python
x = df['city'].value_counts().index
y = df['city'].value_counts().values
plt.bar(x,y,label="real")
plt.xticks(fontproperties=font,size=10)
plt.title("各縣市統計數量",fontproperties=font,size=12)
plt.xlabel("city")
plt.ylabel("number")
plt.savefig(fname="長條圖.png", format="png")
```

# 3. 常用圖表

折線圖 Line plot

```python
x = df['坪數']
plt.plot(x)
plt.show()
```

Scatter plot

```python
x = df['price']
y = df['坪數']
plt.scatter(x,y)
plt.show()
```

圓餅圖 Pie chart

```python
x = df['city'].value_counts()
plt.pie(x)
plt.show()
```

```python
x = df['city'].value_counts()
plt.pie(x,autopct='%1.0f%%')
plt.show()
```

```python
x = df['city'].value_counts()
label = df['city'].value_counts().index
plt.pie(x,autopct='%1.0f%%')
plt.legend(label,prop=font)
plt.show()
```

```python
x = df['city'].value_counts()
label = df['city'].value_counts().index
color = ['#FF3333','#FFAA33','#FFFF00','c','m','y','k','w']
plt.pie(x,autopct='%1.0f%%',colors=color)
plt.legend(label,prop=font)
plt.show()
```

# 4. Seaborn

```python
import seaborn as sns
```

長條圖 Bar chart

```python
sns.countplot(data=df,x='city')
plt.show()
```

```python
sns.countplot(data=df,x='city')
plt.xticks(fontproperties=font,size=10)
plt.title("各縣市統計數量",fontproperties=font,size=12)
plt.xlabel("city")
plt.ylabel("number")
plt.show()
```

折線圖 Line plot

```python
sns.factorplot(data = df, x=df.index, y="坪數")
plt.gcf().set_size_inches(15,5)
plt.show()
```

散步圖 Scatter plot

```python
sns.jointplot(data = df, x="坪數", y="price")
plt.show()
```
