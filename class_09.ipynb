# 4. 訓練資料

Training and testing

```python
X = df_iris.drop("species",axis=1)
y = df_iris['species']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=21)
```

# 5. 預測模型 - 分類 Classification

### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
knn_predict = knn.predict(X_test)
```

```python
pd.DataFrame(list(zip(y_test, knn_predict)),columns=['Measured','Predicted']).head()
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy:',accuracy_score(y_test,knn_predict))
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test, knn_predict))
```

```python
print('Precision:', metrics.precision_score(y_test, knn_predict, average='macro'))
print('Recall:', metrics.recall_score(y_test, knn_predict, average='macro'))
print('F1 score:',metrics.f1_score(y_test, knn_predict, average='macro'))
```

### Support Vector Machine 

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm = SVC(kernel = 'rbf', gamma = 0.2)
```

```python
svm.fit(X_train, y_train)
```

```python
svm_predict = svm.predict(X_test)
```

```python
pd.DataFrame(list(zip(y_test, svm_predict)), columns=['Measured','Predicted']).head()
```

```python
print('Accuracy:',accuracy_score(y_test, svm_predict))
```

# 6. 預測模型 - 分群 Cluster

K-means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 3)
```

```python
kmeans_fit = kmeans.fit(X)
```

```python
kmeans_result = kmeans_fit.labels_
```

```python
from sklearn.metrics import silhouette_score
print('Silhouette:',silhouette_score(X, kmeans_result))
```

```python
import numpy as np
y_c = y.astype('category').cat.codes
y_c = np.array(y_c)
```

```python
plt.scatter(X.values[y_c == 0, 0], X.values[y_c == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X.values[y_c == 1, 0], X.values[y_c == 1, 1], s = 100, c = 'green', label = 'Iris-versicolour')
plt.scatter(X.values[y_c == 2, 0], X.values[y_c == 2, 1], s = 100, c = 'blue', label = 'Iris-virginica')

plt.legend()
plt.title("Real Classification")
plt.show()
```

```python
plt.scatter(X.values[kmeans_result == 0, 0], X.values[kmeans_result == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X.values[kmeans_result == 1, 0], X.values[kmeans_result == 1, 1], s = 100, c = 'green', label = 'Iris-versicolour')
plt.scatter(X.values[kmeans_result == 2, 0], X.values[kmeans_result == 2, 1], s = 100, c = 'blue', label = 'Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeans_fit.cluster_centers_[:, 0], kmeans_fit.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
plt.title("Kmeans Classification")
plt.show()
```
