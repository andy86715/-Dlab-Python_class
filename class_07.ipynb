將上述五種資訊存入五種list裡

```python
a, b, c, d, e = [], [], [], [], []
for i in range(1,11):
    a.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[1]/span[2]').text)
    b.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[2]/div[2]/table/tbody/tr/td[2]/label/span').text)
    c.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[2]/div[2]/table/tbody/tr/td[4]/label/span').text)
    d.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[2]/div[2]/table/tbody/tr/td[5]/label/span[1]').text)
    e.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[2]/div[1]/div/span/span[2]').text)
```

點選換頁按鈕

每抓完10筆資訊即按下一頁的按鈕

```python
a, b, c, d, e = [], [], [], [], []
page = 4
while True:
    for i in range(1,11):
        try:
            a.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[1]/span[2]').text)
            b.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[2]/div[2]/table/tbody/tr/td[2]/label/span').text)
            c.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[2]/div[2]/table/tbody/tr/td[4]/label/span').text)
            d.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[2]/div[2]/table/tbody/tr/td[5]/label/span[1]').text)
            e.append(driver.find_element_by_xpath('//*[@id="main"]/div['+str(i)+']/form/div[2]/div[1]/div/span/span[2]').text)
        except:
            pass
    try:
        driver.find_element_by_xpath('//*[@id="cheapResult-wrapper"]/div[2]/div/span['+str(page)+']/a').click()
        page += 1
    except:
        break
driver.close()
```

zip合併list

```python
information = list(zip(a,b,c,d,e))
```

轉成DataFrame

```python
import pandas as pd
df = pd.DataFrame(information, columns=['航空公司','起飛時間','抵達時間','飛行時間','票價'])
df
```

匯出成Excel

```python
df.to_excel("航班資訊.xlsx")
```
